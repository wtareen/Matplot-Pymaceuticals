```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

mouse_metadata.head()
# Combine the data into a single dataset

# Display the data table for preview

# combined_study_mouse = study_results.set_index('Mouse ID').join(mouse_metadata.set_index('Mouse ID') \
#                                                                , how = 'left')

# combined_study_mouse.head()

# Followiong will reset index from Mouse ID to 0,1,2,3,,,,,,etc.
combined_study_mouse = study_results.set_index('Mouse ID').join(mouse_metadata.set_index('Mouse ID') \
                                                                , how = 'left').reset_index()
combined_study_mouse.head()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
mean1 = combined_study_mouse.groupby('Drug Regimen').mean()['Tumor Volume (mm3)']
median1 = combined_study_mouse.groupby('Drug Regimen').median()['Tumor Volume (mm3)']
var1 = combined_study_mouse.groupby('Drug Regimen').var()['Tumor Volume (mm3)']
std1 = combined_study_mouse.groupby('Drug Regimen').std()['Tumor Volume (mm3)']
sem1 = combined_study_mouse.groupby('Drug Regimen').sem()['Tumor Volume (mm3)']

summaryDataFrame = pd.DataFrame({'Mean': mean1 , 'Median': median1 , 'Variance' : var1, 'STD': std1 , 'SEM': sem1})



# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 

# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

#type(summary_Statistics)
#summary_Statistics
summaryDataFrame
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor 
# volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line

summary_Statistics = combined_study_mouse.groupby('Drug Regimen').agg({'Tumor Volume (mm3)' :\
                                                                       ['mean','median','var','std','sem']})

summary_Statistics
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse

max_TimePoint = combined_study_mouse.groupby('Mouse ID').max()['Timepoint']
max_TimePoint = max_TimePoint.reset_index()
max_TimePoint


# Merge this group df with the original dataframe to get the tumor volume at the last timepoint

# combined_study_mouse = study_results.set_index('Mouse ID').join(mouse_metadata.set_index('Mouse ID') \
#                                                            , how = 'left')
merge_DF = pd.merge(max_TimePoint, combined_study_mouse , how = 'inner' , on = ['Mouse ID','Timepoint'])
merge_DF
```

```python
# Put treatments into a list for for loop (and later for plot labels)

drug_List = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']


# Create empty list to fill with tumor vol data (for plotting)

tumor_Vol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for x in drug_List:
    drug_Vol = merge_DF.loc[merge_DF['Drug Regimen'] == x]['Tumor Volume (mm3)']
    
    tumor_Vol.append(drug_Vol)
    quartiles = drug_Vol.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outlier = drug_Vol.loc[(drug_Vol < lower_bound) | (drug_Vol > upper_bound)]
    
    print(x,outlier)
    
    
    # Locate the rows which contain mice on each drug and get the tumor volumes


    
    # add subset 
    
    
    # Determine outliers using upper and lower bounds
    
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

tumor_Weight = combined_study_mouse.loc[combined_study_mouse["Drug Regimen"] == 'Capomulin'].groupby(['Mouse ID']).agg({'Tumor Volume (mm3)':'mean',\
                                               'Weight (g)':'mean'})
tumor_Weight.head()

plt.subplots(figsize = (10,7))
plt.scatter(tumor_Weight['Weight (g)'].values , tumor_Weight['Tumor Volume (mm3)'].values)
plt.title ('No. of Measurements')
plt.xlabel ('Tumor Volume (mm3)')
_=plt.ylabel( 'Weight')
plt.grid()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
#########
correlation = st.pearsonr(tumor_Weight['Weight (g)'].values,tumor_Weight['Tumor Volume (mm3)'].values)
print(f"The correlation between both factors is {round(correlation[0],2)}")

##############


#####
x_values = tumor_Weight['Weight (g)'].values 
y_values = tumor_Weight['Tumor Volume (mm3)'].values
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Tumor Weight')
plt.ylabel('Tumor Volume')
plt.show()

#####
```
